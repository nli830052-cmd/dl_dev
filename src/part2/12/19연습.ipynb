```python
import torch
x = torch.arange(20).reshape(4,5)

print(x)
print('-------------------------------------')
x = torch.arange(20).reshape(4,-1)
print(x)
```

```python
print(x[1,2])
print(x[2,:])
print(x[:,3])

print(x[1:3,2:4])
mask = x > 10
print(mask)
print(x[mask])
```

```python
# 텐서 이어 붙이기
t1 = torch.arange(6).reshape(2,3)
t2 = torch.arange(6,12).reshape(2,3)

print(t1)
print(t2)

# dim=0 : 행따라 뿥임
cat_dim = torch.cat([t1,t2])
print(cat_dim)

# dim=1 : 열따라 붙ㄱ이기
cat_dim1 = torch.cat([t1,t2],dim=1)
print(cat_dim1)
```

```python
# torch.stack(): 새로운 차원으로 텐서 쌓기
t1 =torch.arange(6).reshape(2,3)
t2 =torch.arange(6,12).reshape(2,3)

stack_dim0 = torch.stack([t1,t2],dim=0)
print(stack_dim0)
```

```python
print(stack_dim0.shape)

stack_dim1 = torch.stack([t1,t2],dim=1)
print(stack_dim1)

stack_dim2 = torch.stack([t1,t2],dim=2)
print(stack_dim2)
```
