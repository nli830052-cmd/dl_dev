```python
# 1.리스트로 텐서 생성
# 0차원 텐서(스칼라)
import torch
from typing import List
scalar_tensor = torch.tensor(7) # 단일값 들어갔으니 스칼라다.
print(scalar_tensor.shape)
```

```python
# 1차원 텐서는 벡 터
list1d:List[int] = [1,2,3,4]
print(list1d)
print(type(list1d))
tensor_1d = torch.tensor(list1d)
print(tensor_1d)

print(tensor_1d.shape)
print(tensor_1d.ndim)

# 2차원 텐서(행렬)
list2d:List[List[int]] = [[1,2,3,4],[5,6,7,8]]
tensor_2d = torch.tensor(list2d)
print(tensor_2d)
print(tensor_2d.shape)

# 3차원 텐서
list3d:List[List[List[int]]] = [
    [[1,2,3],[4,5,6],[7,8,9]]
]
tensor_3d = torch.tensor(list3d)
print(tensor_3d)
print(tensor_3d.shape) # 즉, "1개의 3×3 행렬을 담고 있는 3차원 텐서"라는 뜻이에요.
```

```python
# 3.텐서 속성 확인
# 사용할 장치를 동적으로 설정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device) # - PyTorch 연산 단위에서는 한 번에 한 디바이스만 지정해야 함

#1.정수리스트를 float32타입의 cpu텐서로 생성
data_int = [1,2,3]
tensor_gpu_float32  = torch.tensor(data_int,dtype=torch.float32,device='cuda')

print(tensor_gpu_float32)
print(tensor_gpu_float32.dtype)
```

```python
# Boolean 텐서를 GPU에 생성
data_bool = [True,False,True]
tensor_gpu_bool = torch.tensor(data_bool,device='cuda')
print(tensor_gpu_bool)
print(tensor_gpu_bool.device)
```

```python
# zero랑 onew : 많이쓰임
# 특정크기의 값으로 생성: zeros,ones
zeros_tensor_float32 = torch.zeros(3,4)
print(zeros_tensor_float32)
print(zeros_tensor_float32.dtype)
print(zeros_tensor_float32.device)

zeros_tensor_float321 = torch.zeros(3,4,device='cuda')
print(zeros_tensor_float321.device)

# 2. 2 2 3 크기의 모든 요소가 1인 int8타입의 GPU텐서 생성
ones_tensor_int8 = torch.ones(2,2,3,device='cuda',dtype=torch.int8)
print(ones_tensor_int8)

images_tensor = torch.randn(1,3,256,256)
# print(images_tensor)

# 특정영역지정해서 clip
mask = torch.zeros(256,256,dtype=torch.bool) # 이미지를 덮을 수 있는 마스크를 생성했다
print(mask)

mask[100:150,100:150] = True
```
